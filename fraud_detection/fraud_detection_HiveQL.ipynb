```python
from pyhive import hive
```

```python
# Hive connection parameters
hive_host = "localhost"
hive_port = 10000   

# Connect to Hive
conn = hive.Connection(host=hive_host, port=hive_port, username="root")
cursor = conn.cursor()
```

```python
# Use the database
cursor.execute("USE Transactions")
```

```python
# Detecting unusually high transaction amounts
query1 = """
SELECT *
FROM transactions
WHERE amount > 2328.694382812379
"""
  
cursor.execute(query1)
result1 = cursor.fetchall()
print("Transactions with unusually high amounts:")
for row in result1:
    print(row)
```

```python
# Detecting transactions with customers on the blacklist
query2 = """
SELECT t.* FROM transactions t JOIN blacklist b ON t.merchant_details = b.merchant_name
"""
cursor.execute(query2)
result2 = cursor.fetchall()
print("Transactions with customers on the blacklist:")
for row in result2:
    print(row)
```

```python
# Detecting transactions from unusual locations
query3 = """
SELECT * FROM transactions WHERE location NOT IN (SELECT location FROM customers)
"""
cursor.execute(query3)
result3 = cursor.fetchall()
print("Transactions from unusual locations:")
for row in result3:
    print(row)
```
